```python
import pandas
import matplotlib.pylab as plt
import seaborn as sns
width=0.35
plt.rcParams["figure.figsize"] = [35.00, 19.50]
plt.rcParams["figure.autolayout"] = True
bar_plot1 = sns.barplot(x, rural, data=df, label="Rural",color='orange')
bar_plot2 = sns.barplot(x,urban, data=df, label="Urban",color='blue')
plt.ylim([77435,3588690])
plt.legend(ncol=2, loc="upper right", frameon=True,fontsize=50)
plt.title("Rural v/s Urban",size = 40,fontweight='bold')
plt.xlabel("Dates",size = 30)
plt.show()
```

```python
dose_1=df['dose_1'].tolist()
dose_2=df['dose_2'].tolist()
dose_pd=df['dose_pd'].tolist()
plt.figure(figsize=(37,13), tight_layout=True)
plt.plot(x,dose_1,'o:r',color='r',label='Dose_1',linewidth=2)
#plt.rc('xtick', labelsize=13)
plt.plot(x,dose_2,'o:r',color='b',label='Dose_2',linewidth=2)
#plt.plot(x,total,'o:r',color='g',label='Dose_2',linewidth=2)
#plt.plot(x,dose_pd,color='o',label='Dose_1')
plt.legend(labels=['Dose_1', 'Dose_2'],ncol=2, loc="upper right", frameon=True,fontsize=40)
plt.ylim([16268,903286])
plt.xlabel("Dates",size = 30)
plt.grid()
plt.title("No. of Doses given Each day (Dose_1 and Dose_2)",size = 40,fontweight='bold')
plt.ylabel("No. of Doses given",size = 30)
plt.show()
```

```python
v_doseone=sns.barplot(data=df, x="vaccine_date", y="dose_1",dodge=False)
v_doseone.set_ylabel("Dose_1", fontsize=30)
v_doseone.set_xlabel("Vaccination Date", fontsize=30)
```

```python
v_dosetwo=sns.barplot(data=df, x="vaccine_date", y="dose_2",dodge=False)
v_dosetwo.set_ylabel("Dose_2", fontsize=30)
v_dosetwo.set_xlabel("Vaccination Date", fontsize=30)
```

```python
v_total=sns.barplot(data=df, x="vaccine_date", y="total",dodge=False)
v_total.set_ylabel("Total Doses", fontsize=30)
v_total.set_xlabel("Vaccination Date", fontsize=30)
#v_total.ylim([194595,4270932])
```
