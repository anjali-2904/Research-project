#Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Dataset Used

```python
dataset=pd.read_csv('Covid Survey Form_Second Wave.csv')
```

```python
dataset.head()
```

#Getting insights about the dataset

```python
# Dataset has 278 rows and 19 columns
dataset.shape
```

##Quick summary of dataset

```python
# The describe() function applies basic statistical computations on the dataset like extreme values, 
# count of data points, standard deviation, etc. Any missing value or NaN value is automatically skipped.
dataset.describe()
```

##Columns and their data types

```python
dataset.info()
```

#Handling Missing Values

```python
# Identifying the number of missing values in every column
dataset.isnull().sum()
```

##Dropping 'Timestamp' and 'Name' column as they are irrelevant for us

```python
dataset = dataset.drop(['Timestamp', 'Name'], axis=1)
```

```python
dataset.isnull().sum()
```

# Dropping NaN values

```python
df = dataset.dropna(axis = 0, how ='any')
```

```python
df.shape
```

```python
# We can notice here that dropping NaN values result into us losing data of people who were not infected in the second wave.
# Using this data further will give us biased predictions. Therefore, we stop and visualize region-wise COVID positivity/negativity rate.
df
```

# Data Visualization

```python
# State/UT VS Were you infected in the second wave?
sns.catplot(x="Were you infected in the second wave?", y="State/UT", kind="swarm", data=dataset)
```

```python
# Number of people who were COVID positive in 2nd wave
dataset['Were you infected in the second wave?'].value_counts()
```

```python
# Pie chart: % of people infected in 2nd wave
labels = 'Yes', 'No', 'Maybe'
sizes = [46, 215, 70]
explode = (0.1, 0, 0)  # only "explode" the 1st slice (i.e. 'Yes')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Bar Graph: Number of people who are COVID positive VS State/UT
y = ['Delhi', 'Haryana', 'Maharashtra', 'Odisha', 'Uttar Pradesh', 'Uttarakhand']
 
# Getting values against each value of y
x = [27, 9, 1, 1, 6, 2]
plt.barh(y, x, color='red')
 
# Setting label of y-axis
plt.ylabel("State/UT")
 
# Setting label of x-axis
plt.xlabel("Number of people who are COVID positive")
plt.title("Positivity Rate")
plt.show()
```

```python
# Bar Graph: Number of people who might be COVID positive VS State/UT
y = ['Delhi', 'Haryana', 'Uttar Pradesh', 'West Bengal']
 
# Getting values against each value of y
x = [12, 3, 1, 1]
plt.barh(y, x, color='yellow')
 
# Setting label of y-axis
plt.ylabel("State/UT")
 
# Setting label of x-axis
plt.xlabel("Number of people who might be COVID positive")
plt.title("Positivity Rate")
plt.show()
```

```python
# Bar Graph: Number of people who are COVID negative VS State/UT
y = ['Andhra Pradesh', 'Assam', 'Bihar', 'Chandigarh', 'Delhi', 'Haryana', 'Himachal Pradesh', 'Jharkhand', 'Madhya Pradesh', 'Maharashtra', 'Manipur', 'Odisha', 'Rajasthan', 'Uttar Pradesh', 'Uttarakhand', 'West Bengal']
 
# Getting values against each value of y
x = [1, 1, 9, 1, 104, 30, 2, 2, 3, 1, 2, 1, 6, 47, 4, 1]
plt.barh(y, x, color='green')
 
# Setting label of y-axis
plt.ylabel("State/UT")
 
# Setting label of x-axis
plt.xlabel("Number of people who are COVID negative")
plt.title("Negativity Rate")
plt.show()
```
